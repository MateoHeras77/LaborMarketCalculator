```python
import pyreadstat
import csv

# Lista de años de los que leeremos los archivos
years = list(range(2010, 2024))  # De 2010 a 2023

# Lista de rutas de archivos
file_paths = [
    f"/Users/mateoheras/Downloads/Labour Force Survey {year}/Data/LFS_April_{year}.sav"
    for year in years
]

# Lista donde se guardarán los meta de cada año
meta_list = []

# Lectura de cada archivo .sav
for file_path in file_paths:
    _, meta = pyreadstat.read_sav(file_path)  # Solo nos interesa la metadata, no el DataFrame
    meta_list.append(meta.column_names_to_labels)

# Obtener todos los nombres de las variables (claves de los diccionarios)
keys = set()
for meta in meta_list:
    keys.update(meta.keys())

# Convertimos las claves a una lista y ordenamos para las filas del CSV
keys = sorted(list(keys))

# Generamos las columnas de años
columnas = [str(year) for year in years]

# Crear el archivo CSV
with open('datos_meta.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Escribimos la fila de encabezados (años)
    writer.writerow(['Variable'] + columnas)
    
    # Escribimos los datos por cada variable (clave)
    for key in keys:
        row = [key]  # Iniciamos con el nombre de la variable (clave)
        for meta in meta_list:
            row.append(meta.get(key, ''))  # Añadimos el valor de la descripción o un vacío si no está
        writer.writerow(row)

print("Archivo CSV generado exitosamente.")
```

```python
import pyreadstat
import pandas as pd

# Lista de años de los que leeremos los archivos
years = list(range(2010, 2024))  # De 2010 a 2023

# Lista de rutas de archivos
file_paths = [
    f"/Users/mateoheras/Downloads/Labour Force Survey {year}/Data/LFS_April_{year}.sav"
    for year in years
]

# Lista donde se guardarán los DataFrames
df_list = []

# Lectura de cada archivo .sav y adición de la columna del año
for year, file_path in zip(years, file_paths):
    df, meta = pyreadstat.read_sav(file_path)
    df['year'] = year  # Añadimos la columna del año
    df_list.append(df)  # Añadimos el DataFrame a la lista

# Unimos todos los DataFrames en uno solo
df_total = pd.concat(df_list, ignore_index=True)

# Guardamos el DataFrame combinado en un archivo CSV
df_total.to_csv('LFS_2010_2023_combined.csv', index=False)

print("DataFrames unidos y archivo CSV generado exitosamente.")
```
