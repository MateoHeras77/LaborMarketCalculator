```python
def calculate_zero_nan_percentages(df):
    # Porcentaje de datos que son 0
    zero_percentage = (df == 0).mean() * 100
    # Porcentaje de datos que son NaN
    nan_percentage = df.isna().mean() * 100

    return zero_percentage, nan_percentage
```

```python

zero_percentage, nan_percentage = calculate_zero_nan_percentages(df)

print("Porcentaje de 0 en cada columna:\n", zero_percentage)
print("Porcentaje de NaN en cada columna:\n", nan_percentage)
```
